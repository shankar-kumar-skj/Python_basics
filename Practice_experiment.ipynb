```python
# Checking Whether a Given Number is Even or Odd
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(num," is Even")
else:
    print(num," is Odd")
```

```python
# Demonstration of List and Tuple in Python
my_list = [10,20,30,40]
my_tuple =(10,20,30,40)
print("List elements: ")
for item in my_list:
    print(item)
print("Tuple elements: ")
for tuple in my_tuple:
    print(tuple)
```

```python
# Looping Over a Sequence Using a For Loop
numbers = [10, 20, 30, 40]
for num in numbers:
    print(num)
```

```python
# Countdown Using a While Loop
num = int(input("Enter a number: "))
while num>=0:
    print(num)
    num -=1
```

```python
# Finding the Sum of All Prime Numbers Below Two Million
limit = 2000000
sum_primes = 0
for num in range(2, limit):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        sum_primes += num
print("Sum of primes below two million:", sum_primes)
```

```python
# Finding the Sum of Even-Valued Terms in the Fibonacci Sequence
limit = 4000000  # you can change this value
a, b = 1, 2
even_sum = 0
while a <= limit:
    if a % 2 == 0:
        even_sum += a
    a, b = b, a + b
print("Sum of even Fibonacci numbers:", even_sum)
```

```python
# Checking Whether a Number is Positive, Negative, or Zero
num = float(input("Enter a number: "))
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
# Checking Divisibility of a Number by 5 and 11
num = int(input("Enter a number: "))
if num % 5 == 0 and num % 11 == 0:
    print("The number is divisible by both 5 and 11.")
else:
    print("The number is NOT divisible by both 5 and 11.")
```
